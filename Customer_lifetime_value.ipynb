### Building Model

```python
X=df[['Number of Open Complaints','Number of Policies','Months Since Policy Inception','Months Since Last Claim','Customer Lifetime Value','Monthly Premium Auto']]
```

```python
y=df['Total Claim Amount']
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)# ÄÃ¡nh giÃ¡ há»‡ sá»‘ cháº·n
```

```python
lm.coef_
```

```python
X_train.columns
```

```python
cdf=pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
cdf
```

### Prediction

```python
predictions=lm.predict(X_test)
```
